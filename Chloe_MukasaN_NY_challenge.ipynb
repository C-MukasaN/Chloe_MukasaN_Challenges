```python
# import file : 

import pandas as pd
import numpy as np
data="2018_Central_Park_Squirrel_Census_-_Squirrel_Data.csv"
df = pd.read_csv(data)
df.head()
```

```python
# number of files :

df.shape
```

```python
# total of missing values :

df.isnull().sum()
```

```python
# count the number of nan in each column :

s = df.isnull().apply(sum, axis=0)

# delete columns that have more than 50 missing values : 

for col in df: 
   if s[col] >= 50:  
       del df[col]
print(df.shape)
df.head()
```

```python
df = df.drop(['Lat/Long'], axis=1) # delete the "Lat/Long" column => already x and y, easier to use
```

```python
print(df.shape)
df.head()
```

```python
# save into a .csv file :

smaller_df = df.to_csv('smaller_df.csv', index = False)
```
